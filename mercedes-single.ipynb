```python
import requests
from bs4 import BeautifulSoup
url = "https://fiyat.mercedes-benz.com.tr/model/a-serisi/fiyat/2021"

r = requests.get(url)

soup = BeautifulSoup(r.content,"html.parser")
```

```python
gelen_veri = soup.find_all("div",{"id":"accordionExample"})
```

```python
gelen_veri
```

```python
satirlar = gelen_veri[0]
```

```python
satirlar
```

```python
bilgiler = satirlar.find_all("span")

modelSatiri = satirlar.find_all("div",{"class":"col-lg-2 col-6 acc-headers"})
paketSatiri = satirlar.find_all("div",{"col-lg-2 col-6 acc-headers f-16-set custom-position hodo"})

yakitSatiri = satirlar.find_all("span",{"class":"split-right fuel f-14-set"})
motorSatiri = satirlar.find_all("span",{"class":"split-right engine f-14-set"})
beygirSatiri = satirlar.find_all("span",{"class":"split-right torque f-14-set"})
sanzumanSatiri = satirlar.find_all("span",{"class":"split-right f-14-set"})
fiyatSatiri = satirlar.find_all("span",{"class":"d-block text-center pull-left-md txt-demi f-18"})
```

```python
from termcolor import colored
for fiyatBilgi,yakitBilgi,motorBilgi,beygirBilgi,sanzumanBilgi,modelBilgi,paketBilgi in zip(fiyatSatiri,yakitSatiri,motorSatiri,beygirSatiri,sanzumanSatiri,modelSatiri,paketSatiri):
    fiyat = fiyatBilgi
    yakit = yakitBilgi
    motor = motorBilgi
    beygir = beygirBilgi
    sanzuman = sanzumanBilgi
    model = modelBilgi
    paket = paketBilgi
    print(colored(model.text,"red") if fiyat.text == " TL" else model.text,paket.text,yakit.text,motor.text,beygir.text,sanzuman.text,colored(fiyat.text,"green"))
```
